## Using lamda functions 

```python
bios['height_category'] = bios['height_cm'].apply(lambda x: 'Short' if x < 165 else ('Average' if x < 185 else 'Tall'))
```

```python
bios[['name','height_category']].sample(10)
```

```python
def categorize_athelete(row):
    if row['height_cm'] < 175 and row['weight_kg'] < 75:
        return 'lightweigth baby easy'
    elif row['height_cm'] < 185 or row['weight_kg'] <=85:
        return 'Middleweight'
    else:
        return 'Heavyweight'

bios_new['Category'] = bios.apply(categorize_athelete,axis=1)
```

```python
bios_new[['name','Category']].head(20)
```

## Aggregating Data

```python
bios['born_date'] = pd.to_datetime(bios['born_date'])
```

```python
bios['year_born'] = bios['born_date'].dt.month
bios['month_born'] = bios['born_date'].dt.month
```

```python
# bios.groupby(bios['born_date'].dt.month)['name'].count().sample(10).reset_index().sort_values('name',ascending=False)
bios.groupby([bios['year_born'],bios['month_born']])['name'].count().reset_index().sort_values('name',ascending=False)
```
