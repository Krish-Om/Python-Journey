## Adding / Removing Columns

```python
coffee
```

```python
import numpy as np
coffee['price'] = np.where(coffee['Coffee Type']=='Espresso',3.99,5.99)
```

## Basics Operations 

```python
coffee['revenue'] = coffee['Units Sold'] * coffee['price']
coffee.rename(columns={'price':'rate'},inplace=True)
```

```python
coffee
```

## Handling Null Values

```python
coffee.loc[[1,2],'Units Sold']=np.nan
```

```python
coffee
```

```python
coffee['Units Sold']=coffee['Units Sold'].interpolate()
```

```python
coffee
```

### Aggregating Data

```python
coffee.groupby(['Coffee Type'])['Units Sold'].sum()

coffee.groupby(['Coffee Type'])['Units Sold'].mean()
```

```python
coffee.groupby(['Coffee Type','Day']).agg({'Units Sold':'sum','rate':'mean'}) ## very helpful
```

```python
#Pivot table
pivot = coffee.pivot(columns='Coffee Type',index='Day',values='revenue')
pivot
```

## Advanced Functionalities

* .shift()
* .rank()
* .cumsum() -> cumulative sum
* .rolling()

```python
coffee['yesterday_revenue'] = coffee['revenue'].shift(2)

## for comparing with two days of data
```

```python
coffee
```

```python
coffee['pct_change']= (coffee['revenue']/coffee['yesterday_revenue']) * 100
```

```python
coffee
```

```python
coffee
```

```python
coffee['cumulative_revenue'] = coffee['revenue'].cumsum()
```

```python
min=coffee.cummax()
min
```

```python
latte = coffee[coffee['Coffee Type']=='Latte'].copy()
latte
```

```python
latte['3day'] = latte['Units Sold'].rolling(3).sum()
```
